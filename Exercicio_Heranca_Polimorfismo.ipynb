```python

class Veiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        self.ligado = False

    def ligar(self):
        self.ligado = True
        return f"{self.marca} {self.modelo} está ligado."

    def desligar(self):
        self.ligado = False
        return f"{self.marca} {self.modelo} está desligado."

    def acelerar(self):
        if self.ligado:
            return f"{self.marca} {self.modelo} está acelerando."
        else:
            return f"Não é possível acelerar. {self.marca} {self.modelo} está desligado."

    def __str__(self):
        estado = "ligado" if self.ligado else "desligado"
        return f"{self.marca} {self.modelo} está {estado}."

class Carro(Veiculo): #Herdando de Veiculo
    def __init__(self, marca, modelo, portas):
        super().__init__(marca, modelo)
        self.portas = portas

    def porta_malas(self):
        return f"O porta-malas do {self.marca} {self.modelo} está aberto."

    def acelerar(self):
        if self.ligado:
            return f"O carro {self.marca} {self.modelo} está acelerando muito rápido!"
        else:
            return f"Não é possível acelerar. O carro {self.marca} {self.modelo} está desligado."

    def __str__(self):
        return f"{self.marca} {self.modelo}, {self.portas} portas, está {'ligado' if self.ligado else 'desligado'}."

class Moto(Veiculo): #Herdando de Veiculo
    def __init__(self, marca, modelo, cilindradas):
        super().__init__(marca, modelo)
        self.cilindradas = cilindradas

    def grau(self):
        if self.ligado:
            return f"A moto {self.marca} {self.modelo} está dando grau!"
        else:
            return f"Não é possível dar grau. A moto {self.marca} {self.modelo} está desligada."

    def acelerar(self):
        if self.ligado:
            return f"A moto {self.marca} {self.modelo} está acelerando muito rápido!"
        else:
            return f"Não é possível acelerar. A moto {self.marca} {self.modelo} está desligada."

    def __str__(self):

        return f"{self.marca} {self.modelo}, {self.cilindradas} cilindradas, está {'ligado' if self.ligado else 'desligado'}."

class Eletro(Veiculo):
    def __init__(self, marca, modelo, autonomia):
        super().__init__(marca, modelo)
        self.autonomia = autonomia
        self.carga = 100

    def carregar(self):
        self.carga = 100

        return f"O {self.modelo} está totalmente carregado."

    def autonomia_restante(self):

        return f"O {self.modelo} tem {self.carga}% de carga restante."

class Patins(Eletro):
    def __init__(self, marca, modelo, autonomia):
        super().__init__(marca, modelo, autonomia)

    def acelerar(self):
        if self.ligado and self.carga > 0:
            self.carga -= 10

            return f"O patins elétrico {self.marca} {self.modelo} está acelerando!"
        elif self.ligado and self.carga <= 0:
            return f"O patins {self.marca} {self.modelo} está sem carga. Por favor, carregue-o."
        else:
            return f"Não é possível acelerar. O patins {self.marca} {self.modelo} está desligado."

# Código principal para testar

if __name__ == "__main__":

    print("--- VEÍCULO BASE ---")
    veiculo_base = Veiculo("Generic", "Modelo X")
    print(veiculo_base)
    print(veiculo_base.ligar())
    print(veiculo_base.acelerar())
    print()

    print("--- CARRO (Herança Simples) ---")
    meu_carro = Carro("Chevrolet ","Celta", 4)
    print(meu_carro)
    print(meu_carro.ligar()) #Método Herdado
    print(meu_carro.porta_malas()) #Método Próprio
    print(meu_carro.acelerar()) #Método Sobrescrito
    print()

    print("--- MOTO (Herança Simples) ---")
    minha_moto = Moto("Yamaha", "Mt 09", 847 )
    print(minha_moto)
    print(minha_moto.ligar()) #Método Herdado
    print(minha_moto.grau()) #Método Próprio
    print(minha_moto.acelerar()) #Método Sobrescrito
    print()

    print("--- PATINS ELÉTRICO (Herança Múltipla) ---")
    meu_patins = Patins("Xiaomi", "Mi Electric Scooter 3", 30)
    print(meu_patins)
    print(meu_patins.ligar())               # Método herdado de Veiculo
    print(meu_patins.autonomia_restante())  # Método herdado de Eletro
    print(meu_patins.acelerar())            # Método Sobrescrito
    print(meu_patins.carregar())            # Método herdado de Eletro
    print()

    print("--- POLIMORFISMO ---")
    veiculos = [veiculo_base, meu_carro, minha_moto, meu_patins]

    print("Acelerando todos os veículos:")
    for veiculo in veiculos:
        print(f"{veiculo.acelerar()}")
```
